```python
import pandas,os,re
```

```python
# read all data into pandas.dataframe

df = pandas.read_csv("indoor/0705_indoor_1m.csv")
df2 = pandas.read_csv("indoor/0705_indoor_2m.csv")
```

```python
# extract all filename
def all_indoor_csv_filename():
    return ['indoor/'+path for path in os.listdir('indoor')]

def all_outdoor_csv_filename():
    return ['outdoor/'+path for path in os.listdir('outdoor')]


# extract by distance(1m, 2m, 3m...)
def csv_filename_by_distance(distance:int=1, indoor=True):
    # usage:  csv_filename_by_distance(2,False)  --> outdoor 2m csvfiles
    #              csv_filename_by_distance(2,True)  --> indoor 2m csvfiles
    files = os.listdir('indoor') if indoor else os.listdir('outdoor')
    csv = list(filter(lambda x:(f"{distance}m" in x),files))
    if indoor:
        return ['indoor/'+file for file in csv]
    return ['outdoor/'+file for file in csv]

# extract by date 
def csv_filename_by_date(date:str, indoor=True):
    files = os.listdir('indoor') if indoor else os.listdir('outdoor')
    csv = list(filter(lambda x:(date in x),files))
    if indoor:
        return ['indoor/'+file for file in csv]
    return ['outdoor/'+file for file in csv]


# read csv files into one df
def read_csvs(files):
    
    df = pandas.read_csv(files[0])
    df['distance'] = re.search(r'(\d+)m.csv', files[0]).group(1)
    
    for file in files[1:]:
        df_temp = pandas.read_csv(file)
        df_temp['distance']=re.search(r'(\d+)m.csv', file).group(1)
        df = df.append(df_temp)
    return df
```

```python

# Output the out_door dataframe into csv
files = all_outdoor_csv_filename()
df_outdoor = read_csvs(files)[['Signal strength (dBm)','distance']]
df_outdoor.columns = ['RSS', 'distance'] # change the cols name
df_outdoor['RSS'] = df_outdoor['RSS'].apply(lambda x: x.strip("dBm")) # remove the "dBm" in data

files = all_indoor_csv_filename()
df_indoor = read_csvs(files)[['Signal strength (dBm)','distance']]
df_indoor.columns = ['RSS', 'distance'] # change the cols name
df_indoor['RSS'] = df_indoor['RSS'].apply(lambda x: x.strip("dBm")) # remove the "dBm" in data
```
